```python

#9
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(10,10))
n = 5000
x = np.random.rand(n)
y =  np.random.rand(n)


x0 = 0.5
y0 = 0.5
d=np.sqrt((x-x0)**2 + (y-y0)**2)

r = 0.4

for i in range(len(d)):
    if d[i]<r:
        plt.scatter(x[i],y[i],c='g',marker='x')
    elif d[i]>r:
        plt.scatter(x[i],y[i],c='r',marker='o')


```
